# Part 0

## Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Part 1 
## Data

```python
import pandas_datareader as reader
import datetime
```

```python
import pandas_datareader.data as web
```

```python
start=datetime.datetime(2017,1,1)
end=datetime.date.today()
```

### Reliance

```python
reliance=web.DataReader('RELIANCE.NS','yahoo',start,end)
```

### Cipla

```python
cipla=web.DataReader('CIPLA.NS','yahoo',start,end)
```

### Nifty

```python
nifty=web.DataReader('^NSEI','yahoo',start,end)
```

# Part 2
## Visualising the data

### Opening Price each day

```python
reliance['Open'].plot(label='Reliance',figsize=(12,8),title='Opening Prices')
nifty['Open'].plot(label='Nifty 50')
cipla['Open'].plot(label='Cipla')
plt.legend();
```

### Volume of stock traded each day

```python
reliance['Volume'].plot(label='Reliance',figsize=(12,8),title='Volume Traded')
nifty['Volume'].plot(label='Nifty 50')
cipla['Volume'].plot(label='Cipla')
plt.legend();
```

```python
# What the day of max trade?
# reliance.index[reliance['Volume'].argmax()]
reliance['Volume'].idxmax()
```

### Total money traded in each stock

```python
reliance['Total Traded']=reliance['Volume']*reliance['Open']
cipla['Total Traded']=cipla['Volume']*cipla['Open']
nifty['Total Traded']=nifty['Volume']*nifty['Open']
```

```python
reliance['Total Traded'].plot(label='Reliance',figsize=(12,8),title='Money Traded')
cipla['Total Traded'].plot(label='Cipla')
nifty['Total Traded'].plot(label='Nifty 50')
plt.legend();
```

### Plotting Moving Averages

```python
reliance['MA 50']=reliance['Open'].rolling(50).mean()
nifty['MA 50']=nifty['Open'].rolling(50).mean()
cipla['MA 50']=cipla['Open'].rolling(50).mean()
reliance['MA 200']=reliance['Open'].rolling(200).mean()
nifty['MA 200']=nifty['Open'].rolling(200).mean()
cipla['MA 200']=cipla['Open'].rolling(200).mean()
```

```python
reliance[['Open','MA 50','MA 200']].plot(title='Moving Average Plot of Reliance',figsize=(12,8))
```

### Scatter Matrix

```python
from pandas.plotting import scatter_matrix
```

```python
stonks=pd.concat([reliance['Open'],nifty['Open'],cipla['Open']],axis=1)
stonks.columns=['Reliance Open','Nifty Open','Cipla Open']
```

```python
scatter_matrix(stonks,figsize=(10,10),alpha=0.2,hist_kwds={'bins' :50,'ec':'black'});
```

# Part 3
## Basic Financial Analysis

### Daily Percent Change / Daily Return

```python
reliance['Returns']=(reliance['Close']/reliance['Close'].shift(1))-1
```

```python
reliance['Returns']=reliance['Close'].pct_change(1)
cipla['Returns']=cipla['Close'].pct_change(1)
nifty['Returns']=nifty['Close'].pct_change(1)
```

```python
reliance['Returns'].hist(bins=100,ec='black')
```

```python
cipla['Returns'].hist(bins=100)
```

```python
nifty['Returns'].hist(bins=100,ec='black')
```

#### Stacked on top of each other

```python
reliance['Returns'].hist(bins=100,figsize=(12,8),alpha=0.4,label='Reliance')
cipla['Returns'].hist(bins=100,figsize=(12,8),alpha=0.4,label='Cipla')
nifty['Returns'].hist(bins=100,figsize=(12,8),alpha=0.4,label='Nifty 50')
plt.legend()
```

### Kernel Density Estimation

```python
reliance['Returns'].plot(kind='kde',figsize=(12,8),label='Reliance')
cipla['Returns'].plot(kind='kde',figsize=(12,8),label='Cipla')
nifty['Returns'].plot(kind='kde',figsize=(12,8),label='Nifty 50')
plt.legend()
```
