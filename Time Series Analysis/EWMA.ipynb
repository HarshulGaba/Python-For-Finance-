# Exponential Weighted Moving Average

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
airline=pd.read_csv('airline_passengers.csv',index_col='Month')
airline.dropna(inplace=True)
```

```python
airline.index=pd.to_datetime(airline.index)
airline.info()
```

```python
airline['6 Month SMA']=airline['Thousands of Passengers'].rolling(6).mean()
```

```python
airline['12 Month SMA']=airline['Thousands of Passengers'].rolling(12).mean()
```

```python
airline[['6 Month SMA','12 Month SMA','Thousands of Passengers']].plot(figsize=(12,8))
```

```python
airline['12 Month EWMA']=airline['Thousands of Passengers'].ewm(span=12).mean()
```

```python
airline[['12 Month EWMA','Thousands of Passengers']].plot(figsize=(12,8))
```
