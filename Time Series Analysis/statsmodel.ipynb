```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import statsmodels.api as sm
```

```python
df=sm.datasets.macrodata.load_pandas().data
```

```python
df.set_index(pd.Index(sm.tsa.datetools.dates_from_range('1959q1','2009q3')))
```

```python
df['realgdp'].plot()
```

```python
gdp_cycle,gdp_trend=sm.tsa.filters.hpfilter(df['realgdp'])
```

```python
df['trend']=gdp_trend
```

```python
df[['trend','realgdp']].iloc[-50:].plot()
```
